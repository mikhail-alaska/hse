```python
class GameEngine:
    def __init__(self):
        self.status = 0
        self.counter = 0
        self.current_scene = "scene_start"
        self.choices_made = []
        self.player_name = ""

        self.scenes = {
            "scene_start": {
                "–ü–æ–Ω—è–ª, –ø–∞–ø": "scene_1A",
                "–ù–æ —Ç—É—Ä–Ω–∏—Ä —á–µ—Ä–µ–∑ —Ç—Ä–∏ –Ω–µ–¥–µ–ª–∏!": "scene_1B"
            },

            "scene_1A": {
                "–ò–≥—Ä–∞—Ç—å –Ω–∞ —É—Ä–æ–∫–∞—Ö –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∏": "scene_2A",
                "–¢–µ–æ—Ä–∏—è: —Å–º–æ—Ç—Ä–µ—Ç—å —Å—Ç—Ä–∏–º—ã –∏ –ø–∞—Ç—á–Ω–æ—É—Ç—ã": "scene_2B"
            },

            "scene_1B": {
                "–ò–≥—Ä–∞—Ç—å –Ω–∞ —É—Ä–æ–∫–∞—Ö –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∏": "scene_2C",
                "–¢–µ–æ—Ä–∏—è: —Å–º–æ—Ç—Ä–µ—Ç—å —Å—Ç—Ä–∏–º—ã –∏ –ø–∞—Ç—á–Ω–æ—É—Ç—ã": "scene_2D"
            },

            "scene_2A": {
                "–°–∫–∞–∑–∞—Ç—å –ø—Ä–∞–≤–¥—É –∏ –ø–æ–∫–∞–∑–∞—Ç—å —Ç–µ—Ç—Ä–∞–¥—å —Å —Ç–∞–∫—Ç–∏–∫–∞–º–∏": "scene_3A",
                "–û–±–≤–∏–Ω–∏—Ç—å –í–∞–Ω—å–∫—É –∏–ª–∏ —Å–∏—Å—Ç–µ–º—É": "ending_pudge_mom_basher"
            },

            "scene_2B": {
                "–°–∫–∞–∑–∞—Ç—å –ø—Ä–∞–≤–¥—É –∏ –ø–æ–∫–∞–∑–∞—Ç—å —Ç–µ—Ç—Ä–∞–¥—å —Å —Ç–∞–∫—Ç–∏–∫–∞–º–∏": "scene_3A",
                "–û–±–≤–∏–Ω–∏—Ç—å –í–∞–Ω—å–∫—É –∏–ª–∏ —Å–∏—Å—Ç–µ–º—É": "ending_tp_cancelled"
            },

            "scene_2C": {
                "–°–∫–∞–∑–∞—Ç—å –ø—Ä–∞–≤–¥—É –∏ –ø–æ–∫–∞–∑–∞—Ç—å —Ç–µ—Ç—Ä–∞–¥—å —Å —Ç–∞–∫—Ç–∏–∫–∞–º–∏": "scene_help_from_alina",
                "–û–±–≤–∏–Ω–∏—Ç—å –í–∞–Ω—å–∫—É –∏–ª–∏ —Å–∏—Å—Ç–µ–º—É": "ending_tp_cancelled"
            },

            "scene_2D": {
                "–°–∫–∞–∑–∞—Ç—å –ø—Ä–∞–≤–¥—É –∏ –ø–æ–∫–∞–∑–∞—Ç—å —Ç–µ—Ç—Ä–∞–¥—å —Å —Ç–∞–∫—Ç–∏–∫–∞–º–∏": "scene_streamer_advice",
                "–û–±–≤–∏–Ω–∏—Ç—å –í–∞–Ω—å–∫—É –∏–ª–∏ —Å–∏—Å—Ç–µ–º—É": "ending_tp_cancelled"
            },

            "scene_help_from_alina": {
                "–ü—Ä–∏–Ω—è—Ç—å –ø–æ–º–æ—â—å –ê–ª–∏–Ω—ã": "scene_3C",
                "–°–¥–µ–ª–∞—Ç—å –≤—Å—ë —Å–∞–º–æ–º—É": "scene_3C"
            },

            "scene_streamer_advice": {
                "–°–ª–µ–¥–æ–≤–∞—Ç—å —Å–æ–≤–µ—Ç—É —Å—Ç—Ä–∏–º–µ—Ä–∞": "scene_3A",
                "–ò–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞—Ç—å –∏ —Å–¥–µ–ª–∞—Ç—å –ø–æ-—Å–≤–æ–µ–º—É": "scene_3A"
            },

            "scene_3A": {
                "–ü—Ä–æ—Ä–≤–∞—Ç—å—Å—è –≤ –∏–≥—Ä–æ–≤–æ–π –∫–ª—É–± –Ω–∞ –Ω–æ—á—å": "scene_pre_final_fatigue",
                "–î–æ–≥–æ–≤–æ—Ä–∏—Ç—å—Å—è —Å –æ—Ç—Ü–æ–º": "ending_cinema"
            },

            "scene_3C": {
                "–ü—Ä–æ—Ä–≤–∞—Ç—å—Å—è –≤ –∏–≥—Ä–æ–≤–æ–π –∫–ª—É–± –Ω–∞ –Ω–æ—á—å": "scene_pre_final_fatigue",
                "–î–æ–≥–æ–≤–æ—Ä–∏—Ç—å—Å—è —Å –æ—Ç—Ü–æ–º": "ending_legenda_tolika"
            },

            "scene_pre_final_fatigue": {
                "–ò–≥—Ä–∞—Ç—å –≤—Å—é –Ω–æ—á—å": "ending_fell_asleep",
                "–í—ã—Å–ø–∞—Ç—å—Å—è –∏ —Å–æ–±—Ä–∞—Ç—å—Å—è —É—Ç—Ä–æ–º": "ending_final_like_movie"
            },

            "ending_pudge_mom_basher": {},
            "ending_tp_cancelled": {},
            "ending_cinema": {},
            "ending_legenda_tolika": {},
            "ending_fell_asleep": {},
            "ending_final_like_movie": {}
        }

        self.scene_intro = {
            "scene_start": "–û—Ç–µ—Ü –≤–≤–æ–¥–∏—Ç –ø–∞—Ç—á 7.36: ¬´–ù–∏–∫–∞–∫–æ–≥–æ –∫–æ–º–ø–∞, —Ç–µ–ª–µ—Ñ–æ–Ω ‚Äî –Ω–∞ –ø–æ–ª–∫—É. –ü–æ–Ω—è–ª?¬ª",
        
            "scene_1A": "–¢—ã —Ä–µ—à–∞–µ—à—å –Ω–µ —Å–ø–æ—Ä–∏—Ç—å. –ù–æ –í–∞–Ω—å–∫–∞ —à–µ–ø—á–µ—Ç: ¬´–ï—Å—Ç—å –≤–∞—Ä–∏–∞–Ω—Ç—ã‚Ä¶ –ø—Ä–∞–∫—Ç–∏–∫–∞ –∏–ª–∏ —Ç–µ–æ—Ä–∏—è?¬ª",
            "scene_1B": "–¢—ã —Å–ø–æ—Ä–∏—à—å —Å –æ—Ç—Ü–æ–º. –û–Ω –∑–æ–ª, –Ω–æ —Ç—É—Ä–Ω–∏—Ä –≤—Å—ë –±–ª–∏–∂–µ. –ß—Ç–æ –±—É–¥–µ—à—å –¥–µ–ª–∞—Ç—å –¥–∞–ª—å—à–µ?",
        
            "scene_2A": "–¢—ã —Ç–∞–π–∫–æ–º –∏–≥—Ä–∞–ª –Ω–∞ —É—Ä–æ–∫–∞—Ö –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∏. –£—á–∏—Ç–µ–ª—å –ø–æ—á—Ç–∏ –∑–∞–º–µ—Ç–∏–ª. –°–µ–π—á–∞—Å —Ä–µ—à–∞–µ—Ç—Å—è –≤—Å—ë.",
            "scene_2B": "–¢—ã —É—à—ë–ª –≤ —Ç–µ–æ—Ä–∏—é: –ø–∞—Ç—á–Ω–æ—É—Ç—ã, —Å—Ç—Ä–∏–º—ã, —Å–±–æ—Ä–∫–∏. –ù–æ —Ä–µ–∞–ª—å–Ω–æ–π –ø—Ä–∞–∫—Ç–∏–∫–∏ –Ω–µ—Ç.",
            "scene_2C": "–¢—ã –ø—Ä–æ–¥–æ–ª–∂–∞–µ—à—å —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∞—Ç—å—Å—è, –Ω–µ—Å–º–æ—Ç—Ä—è –Ω–∞ –∑–∞–ø—Ä–µ—Ç. –î–∞–≤–ª–µ–Ω–∏–µ —Ä–∞—Å—Ç—ë—Ç.",
            "scene_2D": "–¢—ã —É–≥–ª—É–±–∏–ª—Å—è –≤ —Ç–µ–æ—Ä–∏—é, –Ω–æ –±–µ–∑ –ø—Ä–∞–∫—Ç–∏–∫–∏ —á—É–≤—Å—Ç–≤—É–µ—à—å –Ω–µ—É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å.",
        
            "scene_help_from_alina": "–ê–ª–∏–Ω–∞ –≤–∏–¥–∏—Ç, —á—Ç–æ —Ç—ã –Ω–∞ –ø—Ä–µ–¥–µ–ª–µ, –∏ –ø—Ä–µ–¥–ª–∞–≥–∞–µ—Ç –ø–æ–º–æ—á—å —Å –ø–æ–¥–≥–æ—Ç–æ–≤–∫–æ–π.",
            "scene_streamer_advice": "–ü–æ–ø—É–ª—è—Ä–Ω—ã–π —Å—Ç—Ä–∏–º–µ—Ä –≥–æ–≤–æ—Ä–∏—Ç: ¬´–ü–æ–±–µ–∂–¥–∞–µ—Ç –Ω–µ —Ç–æ—Ç, –∫—Ç–æ –Ω–µ —Å–ø–∏—Ç, –∞ —Ç–æ—Ç, –∫—Ç–æ –¥—É–º–∞–µ—Ç¬ª.",
        
            "scene_3A": "–î–æ —Ç—É—Ä–Ω–∏—Ä–∞ –æ—Å—Ç–∞–ª–∞—Å—å –æ–¥–Ω–∞ –Ω–æ—á—å. –¢–µ–ø–µ—Ä—å —Ä–µ—à–∞–µ—Ç –Ω–µ —Ç–æ–ª—å–∫–æ —Å–∫–∏–ª–ª, –Ω–æ –∏ –≤—ã–±–æ—Ä.",
            "scene_3C": "–£ —Ç–µ–±—è –µ—Å—Ç—å —á—ë—Ç–∫–∏–π –ø–ª–∞–Ω –æ—Ç –ê–ª–∏–Ω—ã. –û—Å—Ç–∞–ª–æ—Å—å —Ç–æ–ª—å–∫–æ –Ω–µ –≤—Å—ë –∏—Å–ø–æ—Ä—Ç–∏—Ç—å.",
        
            "scene_pre_final_fatigue": "–ù–æ—á—å –ø–µ—Ä–µ–¥ —Ç—É—Ä–Ω–∏—Ä–æ–º. –ì–ª–∞–∑–∞ —Å–ª–∏–ø–∞—é—Ç—Å—è, –Ω–æ —Ä—É–∫–∏ –≤—Å—ë –µ—â—ë —Ç—è–Ω—É—Ç—Å—è –∫ –∫–ª–∞–≤–∏–∞—Ç—É—Ä–µ.",
        
            "ending_pudge_mom_basher": "–¢—ã —Å–≤–∞–ª–∏–ª –≤—Å—ë –Ω–∞ –¥—Ä—É–≥–∏—Ö. –ö–æ–º–∞–Ω–¥–∞ —Ä–∞–∑–≤–∞–ª–∏–ª–∞—Å—å, –∞ —Ç—ã —Å—Ç–∞–ª —à–∫–æ–ª—å–Ω—ã–º –º–µ–º–æ–º.",
            "ending_tp_cancelled": "–¢–µ–±—è –ø–æ–π–º–∞–ª–∏. –¢—É—Ä–Ω–∏—Ä –ø—Ä–æ—à—ë–ª –º–∏–º–æ. –û–±–∏–¥–Ω–æ, –Ω–æ —ç—Ç–æ —Ç–æ–∂–µ –æ–ø—ã—Ç.",
            "ending_cinema": "–¢—ã –¥–æ–≥–æ–≤–æ—Ä–∏–ª—Å—è —Å –æ—Ç—Ü–æ–º. –û–Ω –ø–æ–≤–µ—Ä–∏–ª –≤ —Ç–µ–±—è. –í—ã —Å—ã–≥—Ä–∞–ª–∏ –∫–∞–∫ –∫–æ–º–∞–Ω–¥–∞ ‚Äî –∏ –ø–æ–±–µ–¥–∏–ª–∏.",
            "ending_legenda_tolika": "–¢—ã —Å—ã–≥—Ä–∞–ª –ø–æ –ø–ª–∞–Ω—É –ê–ª–∏–Ω—ã. –£–≤–µ—Ä–µ–Ω–Ω–æ, —Ö–ª–∞–¥–Ω–æ–∫—Ä–æ–≤–Ω–æ ‚Äî –∏ –∑–∞—Ç–∞—â–∏–ª —Ç—É—Ä–Ω–∏—Ä.",
            "ending_fell_asleep": "–¢—ã —É—Å–Ω—É–ª –ø—Ä—è–º–æ –≤–æ –≤—Ä–µ–º—è –¥—Ä–∞—Ñ—Ç–∞. –ö–æ–º–∞–Ω–¥–∞ –≤—ã—à–ª–∞ –±–µ–∑ —Ç–µ–±—è.",
            "ending_final_like_movie": "–¢—ã –≤—ã—Å–ø–∞–ª—Å—è, —Å–æ–±—Ä–∞–ª—Å—è –∏ —Å—ã–≥—Ä–∞–ª –ª—É—á—à–∏–π –º–∞—Ç—á –≤ –∂–∏–∑–Ω–∏. –§–∏–Ω–∞–ª ‚Äî –∫–∞–∫ –≤ –∫–∏–Ω–æ."
        }


    def start(self):
        self.player_name = input("–í–≤–µ–¥–∏—Ç–µ –∏–º—è –∏–≥—Ä–æ–∫–∞: ")
        self.handle_scene("scene_start")

    def handle_scene(self, scene_name):
        self.current_scene = scene_name
        self.counter += 1

        choices = self.scenes.get(scene_name, {})

        print("\n" + self.scene_intro.get(scene_name, "[–û–ø–∏—Å–∞–Ω–∏–µ —Å—Ü–µ–Ω—ã –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç]"))

        if not choices:
            print(f"\n–ö–æ–Ω–µ—Ü –∏–≥—Ä—ã: {scene_name}")
            print(f"–í—ã –ø—Ä–æ—à–ª–∏ –∏–≥—Ä—É –∑–∞ {self.counter} —à–∞–≥–æ–≤.")
            self.save_results()
            self.dfs()
            self.count_endings()
            return

        print("–í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ:")
        options = list(choices.keys())
        for i, option in enumerate(options):
            print(f"{i + 1}: {option}")

        try:
            choice_index = int(input("–í–∞—à –≤—ã–±–æ—Ä (–Ω–æ–º–µ—Ä): ")) - 1
            if 0 <= choice_index < len(options):
                choice_text = options[choice_index]
                self.choices_made.append(choice_text)
                next_scene = choices[choice_text]
                self.current_scene = next_scene
                self.check_next_scene()
            else:
                print("–ù–µ–≤–µ—Ä–Ω—ã–π –Ω–æ–º–µ—Ä. –ü–æ–≤—Ç–æ—Ä–∏—Ç–µ –≤–≤–æ–¥.")
                self.handle_scene(scene_name)
        except ValueError:
            print("–í–≤–µ–¥–∏—Ç–µ —Ç–æ–ª—å–∫–æ —á–∏—Å–ª–æ.")
            self.handle_scene(scene_name)

    def check_next_scene(self):
        if self.current_scene.startswith("ending_"):
            print("\n" + self.scene_intro.get(self.current_scene, "[–û–ø–∏—Å–∞–Ω–∏–µ –∫–æ–Ω—Ü–æ–≤–∫–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç]"))
            print(f"\nüèÅ –ö–æ–Ω–µ—Ü –∏–≥—Ä—ã: {self.current_scene}")
            print(f"üî¢ –í—ã –ø—Ä–æ—à–ª–∏ –∏–≥—Ä—É –∑–∞ {self.counter} —à–∞–≥–æ–≤.")
            self.save_results()
            self.dfs()
            self.count_endings()
        else:
            self.handle_scene(self.current_scene)

    def check_status(self):
        self.start()

    def dfs(self):
        print("\nüß† –ê–Ω–∞–ª–∏–∑ –ø—É—Ç–µ–π (DFS):")
        min_steps = float('inf')

        def dfs_recursive(scene, steps):
            nonlocal min_steps
            if scene.startswith("ending_"):
                min_steps = min(min_steps, steps)
                return
            for next_scene in self.scenes.get(scene, {}).values():
                dfs_recursive(next_scene, steps + 1)

        dfs_recursive("scene_start", 0)

        print(f"üîç –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤ –¥–æ –∫–æ–Ω—Ü–∞: {min_steps}")
        if self.counter > min_steps:
            print(f"–ú–æ–∂–Ω–æ –±—ã–ª–æ –ø—Ä–æ–π—Ç–∏ –±—ã—Å—Ç—Ä–µ–µ –Ω–∞ {self.counter - min_steps} —à–∞–≥(–∞/–æ–≤).")
        else:
            print("–í—ã –ø—Ä–æ—à–ª–∏ —Å–∞–º—ã–º –∫–æ—Ä–æ—Ç–∫–∏–º –ø—É—Ç—ë–º!")

    def count_endings(self):
        print("\nüå≤ –ü–æ–¥—Å—á—ë—Ç –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∫–æ–Ω—Ü–æ–≤–æ–∫...")

        endings = set()

        def recurse(scene):
            if scene.startswith("ending_"):
                endings.add(scene)
                return
            for next_scene in self.scenes.get(scene, {}).values():
                recurse(next_scene)

        recurse("scene_start")
        print(f"üì¶ –í—Å–µ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∫–æ–Ω—Ü–æ–≤–æ–∫: {len(endings)}")
        print("üìå –°–ø–∏—Å–æ–∫ –∫–æ–Ω—Ü–æ–≤–æ–∫:")
        for e in sorted(endings):
            print(f" - {e}")

    def save_results(self):
        with open("results.txt", "a", encoding="utf-8") as f:
            f.write(f"üë§ –ò–≥—Ä–æ–∫: {self.player_name}\n")
            f.write("üìú –°–¥–µ–ª–∞–Ω–Ω—ã–µ –≤—ã–±–æ—Ä—ã:\n")
            for i, choice in enumerate(self.choices_made, 1):
                f.write(f"  {i}. {choice}\n")
            f.write(f"üèÅ –§–∏–Ω–∞–ª: {self.current_scene}\n")
            f.write(f"‚≠ê –û–ø—ã—Ç (—à–∞–≥–∏): {self.counter}\n")
            f.write(f"{'-'*40}\n")

        print("üìù –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ —Ñ–∞–π–ª results.txt.")
```
